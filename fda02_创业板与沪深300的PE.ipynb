```python
# 获取创业板与沪深300成分股2019年1-12月每个交易日的市盈率数据，并展示
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CYB=index_components('399006.XSHE')
HS300=index_components('000300.XSHG')
```

```python
qHS300=query(fundamentals.eod_derivative_indicator.pe_ratio).filter(
fundamentals.eod_derivative_indicator.stockcode.in_(HS300))

qCYB=query(fundamentals.eod_derivative_indicator.pe_ratio).filter(
fundamentals.eod_derivative_indicator.stockcode.in_(CYB))
```

```python
# 获取交易日数据
tradingDates=get_trading_dates('20190101','20191231')
tradingDates
```

```python
d1=tradingDates[0]
d2=tradingDates[1]
```

```python
HS300_d1=get_fundamentals(qHS300, d1, expect_df=True).reset_index()
HS300_d2=get_fundamentals(qHS300, d2, expect_df=True).reset_index()
```

```python
HS300_d12=pd.concat([HS300_d1, HS300_d2])
HS300_d12
```

```python
# 将沪深300交易日的PE数据拼接在一起
fullDataHS300=pd.DataFrame()
for i in tradingDates:
    dailyData=get_fundamentals(qHS300, i, expect_df=True).reset_index()
    fullDataHS300=pd.concat([fullDataHS300, dailyData])
fullDataHS300
```

```python
# 将创业板交易日的PE数据拼接在一起
fullDataCYB=pd.DataFrame()
for i in tradingDates:
    dailyData=get_fundamentals(qCYB, i, expect_df=True).reset_index()
    fullDataCYB=pd.concat([fullDataCYB, dailyData])
fullDataCYB
```

```python
# 用分组求平均功能，计算PE
meanHS300PE=fullDataHS300.groupby('date').mean()
meanCYBPE=fullDataCYB.groupby('date').mean()
```

```python
# 可视化
plt.plot(meanCYBPE, label='创业板')
plt.plot(meanHS300PE, label='沪深300')
plt.legend()
plt.show()
```
