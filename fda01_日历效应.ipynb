```python
# 案例1：日历效应
index_components = index_components('000001.XSHG')
```

```python
all_closes = get_price(index_components,
                       start_date='2000-01-01',
                       end_date='2019-12-31',
                       fields='close')
```

```python
all_closes.head(30)
```

```python
import pandas as pd
import numpy as np
# 移除缺失值大于50个的股票
effective_symbols = pd.DataFrame(all_closes.isna().sum())
effective_symbols.columns = ['NA_count']
effecive_symbol_list = effective_symbols[
    effective_symbols['NA_count'] < 50].index
effective_all_closes = all_closes[effecive_symbol_list]
effective_all_closes
```

```python
# 用平均值填充缺失值
effective_all_closes.isna().sum().sum()
```

```python
effective_all_closes.mean()
```

```python
effective_all_closes = effective_all_closes.fillna(effective_all_closes.mean())
effective_all_closes
```

```python
effective_all_closes.isna().sum().sum()
```

```python
# 频率转化从日价格转为月价格
# 创建日期范围的筛选指标，每月首个交易日
Nmonth_start = pd.date_range('2000-01-01', '2019-12-31', freq='MS').to_list()
Nmonth_start
```

```python
# 处理月首不显交易日的问题，要往后延一天
from datetime import timedelta
for i in range(len(Nmonth_start)):
    while Nmonth_start[i] not in effective_all_closes.index:
        Nmonth_start[i] += timedelta(days=1)
```

```python
Nmonth_start
```

```python
# 创建日期范围的筛选指标，每月最后一个交易日
Nmonth_end = pd.date_range('2000-01-01', '2019-12-31', freq='M').to_list()
Nmonth_end
```

```python
# 处理月末不显交易日的问题，要往前提一天
for i in range(len(Nmonth_end)):
    while Nmonth_end[i] not in effective_all_closes.index:
        Nmonth_end[i] += timedelta(days=-1)
```

```python
Nmonth_end
```

```python
month_start_price = effective_all_closes.loc[Nmonth_start]
month_end_price = effective_all_closes.loc[Nmonth_end]
```

```python
# 不能直接加减，索引要是一致
month_index = month_start_price.index.strftime('%m')
month_start_price.index = month_start_price.index.strftime('%Y%m')
month_end_price.index = month_end_price.index.strftime('%Y%m')
```

```python
month_start_price
```

```python
month_end_price
```

```python
month_return = (month_end_price - month_start_price) / month_start_price
month_return
```

```python
# 二值化，分析
up_or_down = month_return
up_or_down[up_or_down > 0] = 1
up_or_down[up_or_down <= 0] = 0
up_or_down
```

```python
up_or_down_df = pd.DataFrame(up_or_down.mean(axis=1))
up_or_down_df.columns = ['freq of up']
up_or_down_df
```

```python
up_or_down_df['Month No.'] = month_index
up_or_down_df.head(24)
```

```python
x = up_or_down_df.groupby('Month No.').mean()
x
```

```python
x.plot(kind='bar')
```
