```python
# 沪深300基础信息
stock_df=ts.get_hs300s()
print(stock_df)
```

```python
# 所有股票信息
stock_basic=ts.get_stock_basics()
print(stock_basic)
```

```python
# 筛选沪深300的股票
stock_300=stock_basic.loc[stock_df['code'].values]
print(stock_300)
```

```python
X=stock_300.drop(['name', 'industry', 'area'],1)
```

```python
# 使用均值漂移聚类
bandwidth=estimate_bandwidth(X)
ms=MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(X)
```

```python
labels=ms.labels_
for company, label in zip(stock_300['name'], labels):
    print(f'{company}-->{label}')
```

```python
# 使用K邻近算法
kmean=KMeans(n_clusters=12, random_state=0)
kmean.fit(X)
```

```python
labels=kmean.labels_
for company, label in zip(stock_300['name'], labels):
    print(f'{company}-->{label}')
```
